# Wordcloud Example

```python
import wordcloud #pip install wordcloud
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import urllib

from wordcloud import WordCloud, STOPWORDS
```

```python
# Read the whole text.
# The Project Gutenberg EBook of Moby Dick; or The Whale, by Herman Melville
link = "https://www.gutenberg.org/files/2701/2701-0.txt"
urlcontent = urllib.urlopen(link)
text = urlcontent.read()
```

```python
# read the mask image
pie_mask = np.array(Image.open("mobydick.jpg"))
stopwords = set(STOPWORDS)
stopwords.add("and") #add all the stop words
stopwords.add("chapter")
stopwords.add("upon")
```

```python
wc = WordCloud(background_color="white", max_words=50, mask=pie_mask, 
               stopwords=stopwords)
# generate word cloud
wc.generate(text)

# store to file
wc.to_file("mobydickwordcloud.jpg")

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
#plt.imshow(pie_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
